# **Banco de Dados 5: Covid-19 Dataset (GitHub)**


## **Descrição**
O **Covid-19 Dataset** contém dados globais sobre a pandemia de Covid-19, coletados por diversas fontes oficiais, como a Organização Mundial da Saúde (OMS), John Hopkins University e outros. Este dataset fornece informações sobre casos, mortes e recuperação por país e região, bem como métricas relacionadas à propagação da doença.

- **Fonte**: [Our World in Data (GitHub)](https://github.com/owid/covid-19-data)  
- **Número de Registros**: Variável, dependendo da data de atualização, mas normalmente contém mais de 200 registros por país/território.  
- **Número de Atributos**: Aproximadamente 50 colunas, dependendo da versão mais recente do dataset.  

## **Atributos**:
1. `iso_code` - Código do país (ex: `USA`, `BRA`).
2. `continent` - Continente do país (ex: Europa, Ásia, América do Norte).
3. `location` - Nome do país/território.
4. `date` - Data de coleta dos dados.
5. `total_cases` - Número total de casos confirmados de Covid-19.
6. `new_cases` - Número de novos casos confirmados no dia.
7. `total_deaths` - Número total de mortes devido à Covid-19.
8. `new_deaths` - Número de novas mortes no dia.
9. `total_recovered` - Número total de indivíduos recuperados.
10. `new_recovered` - Número de novos recuperados no dia.
11. `active_cases` - Número de casos ativos.
12. `critical_condition` - Número de casos em estado crítico.
13. `total_tests` - Número total de testes realizados.
14. `tests_per_million` - Número de testes realizados por milhão de pessoas.
15. `population` - População total do país.

## **Objetivo**:
Este dataset é utilizado para análises sobre a propagação do coronavírus ao redor do mundo, evolução dos casos, e para entender o impacto da pandemia em diferentes países e continentes. Pode ser usado para modelagem preditiva, visualizações e análises comparativas entre diferentes regiões.

```python
import pandas as pd

# URL do dataset Covid-19
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/latest/owid-covid-latest.csv"

# Carregar o dataset
covid_data = pd.read_csv(url)

# Exibir os primeiros registros
covid_data.head()


```
