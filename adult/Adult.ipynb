# **Banco de Dados 4: Adult Income (Sklearn)**

## **Descrição**
O **Adult Income** (também conhecido como *Census Income*) é um dataset extraído do censo dos Estados Unidos e contém informações demográficas de indivíduos. O objetivo é prever se uma pessoa ganha mais ou menos de $50.000 por ano com base em características como idade, educação, ocupação, etc.

- **Fonte**: [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/adult)  
- **Número de Registros**: 32.561 instâncias  
- **Número de Atributos**: 15 atributos (variáveis explicativas)  
- **Classe (target)**: A variável de classe é se o indivíduo ganha acima ou abaixo de $50.000 por ano. A classe é representada por dois valores: `<=50K` e `>50K`.  

## **Atributos**:
1. `age` - Idade do indivíduo.
2. `workclass` - Tipo de trabalho (ex: privado, público, etc.).
3. `fnlwgt` - Peso final da população (um valor de ponderação).
4. `education` - Nível educacional (ex: Bacharel, Mestrado, etc.).
5. `education-num` - Número de anos de educação.
6. `marital-status` - Estado civil (ex: casado, solteiro).
7. `occupation` - Ocupação do indivíduo (ex: administrativo, técnico, etc.).
8. `relationship` - Relacionamento familiar (ex: casado, não casado, etc.).
9. `race` - Raça do indivíduo.
10. `sex` - Gênero (masculino ou feminino).
11. `capital-gain` - Ganho de capital (valores em dinheiro obtidos com investimentos).
12. `capital-loss` - Perda de capital (valores perdidos com investimentos).
13. `hours-per-week` - Número de horas trabalhadas por semana.
14. `native-country` - País de origem.
15. `class` - Classe de renda (alvo da previsão) - `<=50K` ou `>50K`.

## **Objetivo**:
O objetivo principal desse dataset é construir um modelo preditivo para classificar se uma pessoa ganha acima ou abaixo de $50.000 por ano com base nas variáveis acima.
