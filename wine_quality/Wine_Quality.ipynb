# **Banco de Dados 2: Wine Quality**

## **Descrição**

O **Wine Quality** dataset contém informações sobre as características químicas de vinhos e a qualidade atribuída a eles. É amplamente utilizado para problemas de classificação ou regressão, com o objetivo de prever a qualidade do vinho com base em suas propriedades.

- **Fonte**: [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/wine+quality)  
- **Número de Registros**: 1.599 instâncias  
- **Número de Atributos**: 12 atributos

## **Atributos**:
1. `fixed acidity` - Acidez fixa do vinho.
2. `volatile acidity` - Acidez volátil do vinho.
3. `citric acid` - Quantidade de ácido cítrico no vinho.
4. `residual sugar` - Quantidade de açúcar residual.
5. `chlorides` - Quantidade de cloreto no vinho.
6. `free sulfur dioxide` - Quantidade de dióxido de enxofre livre.
7. `total sulfur dioxide` - Quantidade total de dióxido de enxofre.
8. `density` - Densidade do vinho.
9. `pH` - pH do vinho.
10. `sulphates` - Quantidade de sulfatos no vinho.
11. `alcohol` - Teor alcoólico do vinho.
12. `quality` - Qualidade do vinho (escala de 0 a 10, sendo 0 a pior e 10 a melhor qualidade).

## **Objetivo**:
O objetivo deste dataset é prever a qualidade do vinho (variável `quality`) com base nas características químicas, como acidez, açúcar, pH, entre outros.


```python
import pandas as pd

# URL do dataset Wine Quality (exemplo: vinhos brancos)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"

# Carregar o dataset
wine = pd.read_csv(url, sep=";")

# Exibir os primeiros registros
wine.head()
```
