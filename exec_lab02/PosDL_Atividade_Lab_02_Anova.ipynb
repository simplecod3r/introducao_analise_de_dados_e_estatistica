### 🧠 **Nota Metodológica (Importante)**
A ANOVA requer uma **variável dependente numérica** e uma **variável independente categórica**.  
No conjunto *Wine Quality*, originalmente as duas variáveis (`alcohol` e `quality`) são numéricas.  
Portanto, para aplicar corretamente a ANOVA One-Way, utilizaremos:

- **Dependente:** `alcohol` (contínua)  
- **Fator:** `quality` (categórica)

Caso desejado, também é possível inverter a análise (tratando `quality` como contínua e categorizando `alcohol`), mas nesse caso deve-se justificar a decisão no relatório.
