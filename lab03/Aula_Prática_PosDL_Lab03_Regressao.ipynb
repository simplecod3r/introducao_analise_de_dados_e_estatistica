# Analise de Regressão

### **Análise de Regressão**

A análise de regressão é uma ferramenta estatística poderosa utilizada para entender a relação entre uma variável dependente (ou resposta) e uma ou mais variáveis independentes (ou preditores). O objetivo principal da regressão é modelar essa relação para prever valores da variável dependente com base nas variáveis independentes, bem como para identificar padrões e tendências nos dados.

Existem diferentes tipos de modelos de regressão, sendo os mais comuns a **regressão linear simples** e a **regressão linear múltipla**.

#### Regressão Linear Simples

A **regressão linear simples** é utilizada quando há uma relação linear entre uma variável dependente e uma única variável independente.

A equação da reta é dada por:

$$
Y = \beta_0 + \beta_1 X + \epsilon
$$

Onde:
- \(Y\) é a variável dependente  
- \(X\) é a variável independente  
- \(\beta_0\) é o intercepto  
- \(\beta_1\) é o coeficiente de regressão  
- \(\epsilon\) é o erro residual  


#### **Regressão Linear Múltipla**
Quando há mais de uma variável independente, a **regressão linear múltipla** entra em cena. Esse modelo generaliza a regressão linear simples e pode ser utilizado para analisar a relação entre uma variável dependente e várias variáveis independentes. A equação do modelo de regressão linear múltipla é dada por:

$$
[
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_nX_n + \epsilon
]$$

Onde:
- \(Y\) é a variável dependente,
- \(X_1, X_2, ..., X_n\) são as variáveis independentes,
- \(\beta_1, \beta_2, ..., \beta_n\) são os coeficientes de regressão associados a cada variável independente,
- \(\beta_0\) é o intercepto,
- \(\epsilon\) é o erro residual.

A regressão linear múltipla permite capturar relações mais complexas, onde várias variáveis podem influenciar a variável dependente simultaneamente.

#### **Aplicações**
A análise de regressão é amplamente utilizada em diversas áreas, como economia, saúde, marketing e ciências sociais. Por exemplo, pode-se usar a regressão para prever vendas com base em gastos com publicidade (regressão linear simples) ou para prever o preço de imóveis com base em várias características, como área, número de quartos e localização (regressão linear múltipla).

#### **Importância**
Compreender a análise de regressão é fundamental para análise preditiva, tomada de decisões e para construir modelos explicativos robustos. A análise de regressão não só permite prever um valor, mas também entender quais fatores mais influenciam o comportamento da variável dependente, proporcionando insights valiosos.

## Regressão Linear Simple

```python
X = iris["petal_length"]
y = iris["petal_width"]

# A função add_constant do pacote statsmodels é usada para adicionar uma coluna de 1s à matriz X.
# Isso é necessário porque, em um modelo de regressão linear, queremos estimar um termo de interceptação (β0),
# que representa o valor da variável dependente (y) quando todas as variáveis independentes (X) são iguais a zero.
# O termo de interceptação é importante para ajustar o modelo e garantir que ele tenha uma "base" de previsão,
# mesmo quando as variáveis independentes não possuem valor (por exemplo, quando o comprimento da pétala é 0,
# o modelo ainda precisa de uma referência para a largura da pétala).
# Essa coluna de 1s corresponde à constante (β0) na equação da reta de regressão:
# y = β0 + β1 * X.

# Vale destacar que o termo de interceptação pode não ter uma interpretação prática em todos os contextos.
# Por exemplo, no caso de prever o preço de uma casa em função de seu tamanho, o tamanho zero não faz sentido,
# pois é impossível ter uma casa com tamanho zero. Assim, o termo de interceptação no modelo de regressão não
# teria uma interpretação prática direta nesse contexto, mas ainda é necessário para o ajuste do modelo.

# Agora, adicionamos a constante à matriz X.
X = sm.add_constant(X)

# Ajustamos o modelo de regressão linear simples usando o método OLS (Ordinary Least Squares).
# OLS ajusta a reta de regressão que minimiza a soma dos quadrados dos erros (resíduos),
# ou seja, a diferença entre os valores observados e os valores previstos pelo modelo.
model = sm.OLS(y, X).fit()  # Ajuste do modelo de regressão linear

# Exibimos o resumo do modelo ajustado, que fornece informações sobre os coeficientes,
# erros padrão, valores t e p-valores, entre outras métricas úteis para avaliar a qualidade do modelo.
print(model.summary())
```

**Explicação do modelo:**

* O que é add_constant(X)? A função sm.add_constant(X) é usada para adicionar uma coluna de 1s à matriz X. Isso é fundamental para que o modelo de regressão tenha um intercepto (termo constante), permitindo que ele ajuste a equação de regressão com um ponto de referência, mesmo quando todas as variáveis independentes são zero.

* Interpretação do termo de interceptação: O intercepto (β0) é o valor da variável dependente (no caso, "petal_width") quando todas as variáveis independentes (como "petal_length") são iguais a zero. Embora este termo seja matematicamente necessário, em muitos contextos, como o preço de uma casa baseado no tamanho, não é realista que a variável independente tenha valor zero. Nesse caso, o intercepto pode não ter uma interpretação prática direta, mas ainda é importante para o modelo.

* Modelo OLS (Ordinary Least Squares): O método de ajuste utilizado, OLS, busca minimizar a soma dos quadrados dos erros, ou seja, ele tenta encontrar a reta de regressão que melhor se ajusta aos dados, minimizando as diferenças entre os valores reais de "petal_width" e os valores previstos pelo modelo.

* model.summary(): O método summary() exibe um conjunto detalhado de informações sobre o modelo ajustado. Isso inclui os coeficientes de regressão, erros padrão, estatísticas t, p-valores e outros indicadores importantes para avaliar a qualidade e significância do modelo.

## Regressão Linear Múltipla

A Regressão Múltipla é uma técnica estatística utilizada para modelar a relação entre uma variável dependente e duas ou mais variáveis independentes. É uma extensão da Regressão Linear Simples, que envolve apenas uma variável independente. Na Regressão Múltipla, a relação entre a variável dependente e as variáveis independentes é modelada por uma equação linear que inclui vários coeficientes, um para cada variável independente.

A diferença é que na regressão múltipla, ao adicionarmos mais variáveis independentes, a linha reta passa a ser um hiperplano que passa através do espaço de todas as variáveis independentes. Segue a mesma logica de implementação da Regressão Linear Simple só que adiciona mais de uma variavel

```python
# Definindo as variáveis independentes (X) e dependente (y)

X = iris[["petal_length", "sepal_length"]]  # Seleciona as variáveis independentes

y = iris["petal_width"]  # Variável dependente

# Adicionando uma constante à matriz X para o cálculo do intercepto
X = sm.add_constant(X)

# Ajustando o modelo de regressão linear múltipla
model = sm.OLS(y, X).fit()

# Exibindo o resumo dos resultados do modelo
print(model.summary())
```

* Resíduos fora da faixa esperada:

Expectativa: Os resíduos devem estar próximos de zero e dentro de uma faixa razoável em relação aos valores previstos.
Interpretação: Se houver muitos resíduos grandes (tanto positivos quanto negativos) fora da faixa normal, isso pode indicar a presença de outliers (valores extremos). Esses outliers podem ter um impacto significativo no modelo, e pode ser necessário investigá-los separadamente.

Expectativa: A linha de resíduos deve se concentrar em torno de zero sem tendência clara.
Interpretação: Se a linha de resíduos estiver bem centrada em zero, isso sugere que o modelo está fazendo boas previsões em média. # Correlação e Casualidade

## 1. Conceito

### Correlação
A **correlação** mede o grau de associação entre duas variáveis, indicando se elas variam juntas e em que direção (positiva ou negativa).  
É importante destacar que **correlação não implica causalidade**.

A correlação mais comum é a **correlação de Pearson**, que assume valores entre:
- **+1**: correlação positiva perfeita  
- **0**: ausência de correlação linear  
- **−1**: correlação negativa perfeita  

### Causalidade
A **causalidade** ocorre quando a variação de uma variável **provoca diretamente** a variação de outra.  
Para afirmar causalidade, não basta observar associação estatística; é necessário atender a critérios como:
- precedência temporal,
- mecanismo causal plausível,
- controle de variáveis de confusão.

Na prática, a maioria das análises em ciência de dados identifica **correlação**, e não causalidade.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset Iris
iris = sns.load_dataset("iris")

# Selecionar variáveis
X = iris["petal_length"]
y = iris["petal_width"]
```

## 2. Correlação na prática

Vamos analisar a correlação entre as variáveis **petal_length** e **petal_width** utilizando o coeficiente de correlação de Pearson.


```python
# Matriz de correlação
corr = iris[["petal_length", "petal_width"]].corr()
corr
```

O valor obtido representa o grau de associação linear entre as duas variáveis.  
Valores próximos de 1 indicam forte correlação positiva.


```python
# Gráfico de dispersão
sns.scatterplot(x=X, y=y)
plt.title("Dispersão entre Petal Length e Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
# Gráfico com reta de regressão (visualização da correlação)
sns.regplot(
    x=X,
    y=y,
    line_kws={"color": "red", "lw": 3, "alpha": 0.7}
)

plt.title("Correlação entre Petal Length e Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

## 3. Correlação ≠ Causalidade

Apesar da forte correlação observada entre **petal_length** e **petal_width**, não é possível afirmar que o comprimento da pétala **causa** o aumento da largura.

Ambas as variáveis podem estar sendo influenciadas por um **terceiro fator**, como a espécie da flor.


```python
# Analisando uma possível variável de confusão
sns.boxplot(
    data=iris,
    x="species",
    y="petal_length"
)
plt.title("Petal Length por Espécie")
plt.show()
```

O gráfico acima mostra que a espécie da flor influencia o comprimento da pétala, reforçando que a correlação observada não implica causalidade direta.


## 4. Exemplo de Causalidade

### 1. Exemplo conceitual (clássico e intuitivo)



#### Uso de fertilizante e crescimento da planta

- **Causa (X):** quantidade de fertilizante aplicada  
- **Efeito (Y):** crescimento da planta  

Se realizarmos um **experimento controlado**, no qual:
- as plantas são divididas aleatoriamente em grupos,
- cada grupo recebe uma quantidade diferente de fertilizante,
- todas as demais condições são mantidas constantes (luz, água e solo),

então qualquer diferença observada no crescimento pode ser atribuída **diretamente ao fertilizante**.

**Neste caso, podemos afirmar causalidade**, pois:
- há precedência temporal (fertilizante → crescimento),
- existe um mecanismo biológico conhecido,
- variáveis de confusão foram controladas.


### 2. Exemplo prático de causalidade (experimento simulado)

Agora vamos simular um experimento causal em Python, no qual o tratamento é atribuído
aleatoriamente, caracterizando um cenário de inferência causal.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Reprodutibilidade
np.random.seed(42)

# Número de observações
n = 200

# Tratamento aleatório (0 = controle, 1 = tratamento)
tratamento = np.random.binomial(1, 0.5, n)


resultado = 10 + 5 * tratamento + np.random.normal(0, 2, n)

df_causal = pd.DataFrame({
    "tratamento": tratamento,
    "resultado": resultado
})

df_causal.head()
```

No código acima:
- a variável **tratamento** é atribuída aleatoriamente, simulando um experimento controlado,
- o valor `5` representa o **efeito causal verdadeiro** do tratamento sobre o resultado,
- o termo aleatório representa o ruído natural dos dados.


```python
# Visualização do efeito causal
sns.boxplot(
    data=df_causal,
    x="tratamento",
    y="resultado"
)

plt.title("Efeito Causal do Tratamento sobre o Resultado")
plt.xlabel("Tratamento (0 = Controle, 1 = Tratamento)")
plt.ylabel("Resultado")
plt.show()
```

O gráfico evidencia que o grupo que recebeu o tratamento apresenta,
em média, valores maiores do resultado em comparação ao grupo de controle.

Como o tratamento foi atribuído de forma aleatória e as demais condições foram controladas,
a diferença observada pode ser interpretada como **efeito causal**.


### 3. Relação com Regressão Linear

A regressão linear, seja **simples** ou **múltipla**, ajusta modelos do tipo:

- Regressão Linear Simples

$$
Y = \beta_0 + \beta_1 X + \epsilon
$$

- Regressão Linear Múltipla

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_p X_p + \epsilon
$$

Em ambos os casos, o modelo:
- explora **associações estatísticas (correlação)** entre as variáveis,
- é amplamente utilizado para **previsão**,
- **não garante inferência causal** por si só.

A presença de múltiplas variáveis explicativas não implica, necessariamente,
controle adequado de variáveis de confusão.
Interpretar os coeficientes $\beta_j$ como efeitos causais requer hipóteses adicionais,
como ausência de confundimento, precedência temporal e, idealmente, um desenho experimental adequado.


> Mesmo em regressões múltiplas, os coeficientes estimados representam associações condicionais,
> não efeitos causais garantidos.

## 5. Conclusão


- Correlação mede associação, não causa  
- Regressão linear utiliza correlação para modelagem preditiva  
- Relações fortes podem ser explicadas por variáveis ocultas  
- Inferência causal exige métodos específicos e controle de confundidores  

Portanto, ao analisar dados observacionais, a conclusão correta é identificar **associação estatística**, e não causalidade.
