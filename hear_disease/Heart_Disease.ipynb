# **Banco de Dados 3: Heart Disease**

## **Descrição**

O **Heart Disease** dataset contém dados médicos de pacientes e é utilizado para prever a presença ou ausência de doenças cardíacas. Este conjunto de dados é comumente usado para modelos de aprendizado supervisionado, como classificação binária, para prever a probabilidade de um paciente ter doença cardíaca.

- **Fonte**: [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Heart+Disease)  
- **Número de Registros**: 303 instâncias  
- **Número de Atributos**: 14 atributos

## **Atributos**:
1. `age` - Idade do paciente.
2. `sex` - Sexo do paciente (0 = feminino, 1 = masculino).
3. `cp` - Tipo de dor no peito (4 tipos: 1 = angina típica, 2 = angina atípica, 3 = dor sem angina, 4 = sem dor).
4. `trestbps` - Pressão arterial em repouso (em mm Hg).
5. `chol` - Colesterol sérico (em mg/dl).
6. `fbs` - Glicemia em jejum (0 = menor que 120 mg/dl, 1 = maior que 120 mg/dl).
7. `restecg` - Resultado do eletrocardiograma em repouso (0 = normal, 1 = com anomalia, 2 = hipertrofia ventricular esquerda).
8. `thalach` - Frequência cardíaca máxima atingida.
9. `exang` - Angina induzida por exercício (0 = não, 1 = sim).
10. `oldpeak` - Depressão do segmento ST induzida por exercício relativo ao repouso.
11. `slope` - Inclinação do segmento ST no teste de exercício.
12. `ca` - Número de vasos principais coloridos pela fluoroscopia (0 a 3).
13. `thal` - Tipo de talassemia (3 tipos: 3 = normal, 6 = deficiente, 7 = reversível).
14. `target` - Presença de doença cardíaca (0 = não, 1 = sim).

## **Objetivo**:
O objetivo deste dataset é prever se um paciente tem ou não a doença cardíaca com base nas variáveis médicas fornecidas, como idade, colesterol, pressão arterial, entre outros.


```python
import pandas as pd

# URL do dataset Heart Disease
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"

# Definir nomes de colunas (específicos para este dataset)
columns = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach",
    "exang", "oldpeak", "slope", "ca", "thal", "target"
]

# Carregar o dataset
heart_disease = pd.read_csv(url, header=None, names=columns, na_values="?")

# Exibir os primeiros registros
heart_disease.head()
```
