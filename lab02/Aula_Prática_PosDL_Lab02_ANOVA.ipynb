# **Análise de Variância (ANOVA) em Python**

A análise de variância (ANOVA) é uma técnica estatística utilizada para determinar se há diferenças significativas entre as médias de dois ou mais grupos de dados. Em Python, podemos realizar ANOVA com a ajuda das bibliotecas statsmodels.

## **Iris Dataset**

Vamos considerar o conjunto de dados classico **Iris** muito utilizado para fins de aprendizado de máquina e análise de dados. Ele contém informações sobre três espécies diferentes de plantas Iris (Iris setosa, Iris virginica e Iris versicolor) e as medidas de quatro características de suas flores: comprimento e largura das pétalas e das sépalas.

O dataset contém 150 observações, sendo 50 para cada espécie de planta Iris. As medidas foram realizadas pelo botânico britânico Ronald Fisher em 1936, e o conjunto de dados é frequentemente utilizado para fins de classificação de espécies de flores Iris com base em suas medidas.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import seaborn as sns

# Carregamos os dados com a lib seaborn
iris = sns.load_dataset('iris')
```

```python
iris.head(5)
```

```python
iris.describe()
```

## Exemplo de código de ANOVA ONE-WAY

**Por que ANOVA one-way?**

A ANOVA one-way é usada quando:

* Há uma única variável independente (ou fator):

Neste caso, species é a única variável independente com três categorias.

* A variável dependente é contínua:

A variável sepal_length atende a este critério.

* O objetivo é comparar as médias entre os grupos:

A ANOVA testa se as médias de sepal_length para as três espécies são iguais ou não.

O objetivo deste estudo é determinar se a média do comprimento da sépala (sepal_length) difere entre as três espécies de flores do conjunto de dados Iris.

Para isso, utilizamos a Análise de Variância (ANOVA) One-Way, que permite comparar as médias de uma variável dependente contínua entre três ou mais grupos categóricos. Neste caso:

* Variável dependente: sepal_length (comprimento da sépala).
* Variável independente: species (espécie de flor: setosa, versicolor, virginica).

**Etapas do Processo**

1. Ajuste do Modelo ANOVA

O modelo ANOVA é ajustado utilizando a função ols() da biblioteca statsmodels.formula.api. Essa função realiza a regressão linear necessária para calcular os componentes da variância entre os grupos.

```python
model = ols('sepal_length ~ species', data=iris).fit()
```

2. Execução da ANOVA

Com o modelo ajustado, executamos a ANOVA usando a função anova_lm() da biblioteca statsmodels.stats.anova.

Aqui, utilizamos o argumento type=2, que indica o uso da soma de quadrados tipo II, recomendada quando os tamanhos dos grupos são diferentes.

```python
from statsmodels.stats.anova import anova_lm
anova_table = anova_lm(model, type=2)
```

3. **Interpretação dos Resultados**

A tabela resultante apresenta os seguintes componentes principais:

* df: Graus de liberdade para cada fonte de variação.
* sum_sq: Soma dos quadrados (variância explicada ou residual).
* mean_sq: Média dos quadrados (variância por grau de liberdade).
* F: Estatística F, que indica a razão entre a variância explicada e a variância residual.
*PR(>F): Valor-p, que indica se a diferença entre as médias é estatisticamente significativa

```python
print(anova_table)
```

**Explicação da Tabela `anova_table`**

A tabela `anova_table` fornece informações sobre a decomposição da variância. Vamos entender cada coluna:

- **df (graus de liberdade)**: Representa a quantidade de informações independentes disponíveis para estimar os parâmetros.  
  - **species**: 2 graus de liberdade, pois há 3 categorias e utilizamos o cálculo:  
    
    Número de categorias - 1 = 3 - 1 = 2
  - **Residual**: 147 graus de liberdade restantes (\(150\) amostras - \(3\) categorias).

- **sum_sq (soma de quadrados)**: Representa a variância explicada (`species`) e a não explicada (`residuals`).

- **mean_sq (média dos quadrados)**: É a soma de quadrados dividida pelos graus de liberdade.

- **F (estatística F)**: Mede a proporção da variância explicada pela variável independente em relação à variância residual.  
  Um valor alto indica uma diferença significativa entre os grupos.

- **PR(>F)**: P-valor associado à estatística F.  
  Um valor muito pequeno (como \(1.67 \times 10^{-31}\)) indica evidências para rejeitar a hipótese nula, ou seja, as médias das espécies são significativamente diferentes.


## Post Hoc

Quando a ANOVA One-Way detecta diferenças significativas entre os grupos (com base no p-valor associado à estatística F), como é o caso aqui (p<0.05), é necessário realizar um teste post hoc. O teste post hoc ajuda a identificar especificamente quais grupos são diferentes entre si.

Neste caso, a análise ANOVA indica que existe uma diferença significativa entre as médias do comprimento da sépala (sepal_length) para as espécies, mas não informa entre quais espécies essas diferenças ocorrem. O post hoc resolve essa questão.

**Como fazer o post hoc?**

Um dos testes mais comuns para isso é o Teste de Tukey (Tukey HSD), que compara todas as combinações de pares de grupos, ajustando os valores-p para múltiplas comparações.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Aplicando o Teste de Tukey
posthoc = pairwise_tukeyhsd(endog=iris['sepal_length'], groups=iris['species'], alpha=0.05)

# Exibindo os resultados
print(posthoc)
```

**Interpretação dos Resultados:**

O Teste de Tukey fornecerá uma tabela com:

* Os pares de grupos comparados.
* A diferença média entre os grupos.
* O intervalo de confiança.
* Um indicador se a diferença é significativa (
p<0.05).

Isso permitirá determinar quais pares de espécies (ex.: setosa vs. versicolor, versicolor vs. virginica, etc.) apresentam diferenças significativas no comprimento médio da sépala.

Interpretação:

* setosa vs versicolor: Diferença média de +0.93, com p<0.05. Isso significa que o comprimento médio da sépala em setosa é significativamente maior do que em versicolor.

* setosa vs virginica: Diferença média de +1.582, com p<0.05. O comprimento médio da sépala em setosa também é significativamente maior do que em virginica.

* versicolor vs virginica: Diferença média de +0.652, com p<0.05. O comprimento médio da sépala em versicolor é significativamente maior do que em virginica.

* Se meandiff > 0, significa que a média da primeira categoria (group1) é maior do que a média da segunda categoria (group2).

* Se meandiff < 0, significa que a média da primeira categoria (group1) é menor do que a média da segunda categoria (group2).

## Titanic Dataset

Vamos utilizar o conjunto de dados Titanic, que contém informações sobre os passageiros, como idade, sexo, classe, tarifa, entre outras variáveis. Para este exemplo de ANOVA Two-Way, analisaremos as seguintes variáveis:

**Variável dependente:**

* fare: Tarifa paga pelo passageiro (variável numérica contínua).

**Variáveis independentes (fatores):**

* class: Classe do passageiro (variável categórica com 3 níveis: primeira, segunda e terceira classe).
* sex: Sexo do passageiro (variável categórica com 2 níveis: masculino e feminino).

**O objetivo da análise é determinar:**

* Se a tarifa média varia entre as diferentes classes de passageiros.

* Se a tarifa média varia entre os sexos (masculino e feminino).

* Se há uma interação significativa entre as variáveis class e sex na determinação da tarifa média.

## Exemplo de código do ANOVA TWO-WAY

Para realizar um exemplo de ANOVA Two-Way, podemos utilizar um conjunto de dados que contenha duas variáveis categóricas e uma variável numérica contínua. Neste caso, vamos utilizar o conjunto de dados Titanic, que está disponível na biblioteca Seaborn do Python.

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(5)
```

```python
# Mudamos o nome da coluna 'class' para 'classe' para evitar conflito com a palavra reservada 'class' do Python
titanic.rename(columns= {'class': 'classe'}, inplace=True)

# Criamos o modelo ANOVA Two-Way, onde a variável dependente é 'fare' (tarifa)

# E as variáveis independentes são 'classe' e 'sex', incluindo a interação entre elas

model = ols('fare ~ classe + sex + classe*sex', data=titanic).fit()

# Executamos a ANOVA com a decomposição dos quadrados tipo 2

anova_table = sm.stats.anova_lm(model, typ=2)

# Exibimos a tabela de resultados
print(anova_table)
```

**Explicação do Código:**

* Renomeação da Coluna:

Você renomeia a coluna class para classe, pois class é uma palavra reservada no Python e pode causar problemas.

* Modelo ANOVA:

O modelo ols('fare ~ classe + sex + classe*sex', data=titanic) ajusta um modelo de ANOVA Two-Way. O que está sendo feito aqui é:
fare: A variável dependente, que é a tarifa paga pelo passageiro.

classe e sex: São as variáveis independentes. Aqui estamos considerando os efeitos principais de cada variável.

classe*sex: A interação entre as variáveis classe e sex, o que é essencial em uma ANOVA Two-Way.

* ANOVA:

A função sm.stats.anova_lm(model, typ=2) realiza a análise de variância com a decomposição tipo 2, que é comum quando temos mais de um fator (no caso, classe e sex).

O tipo 2 é recomendado quando os fatores têm diferentes números de níveis e queremos avaliar os efeitos principais e as interações.
Resultados:

O código print(anova_table) vai exibir a tabela de ANOVA com os valores de soma de quadrados, graus de liberdade, estatísticas F e p-valores. Esses valores serão usados para interpretar os efeitos principais e as interações.

### Interpretando os resultados

**Interpretação Final:**

* Efeitos principais:

Tanto a variável classe quanto a variável sex têm efeitos significativos sobre a tarifa média paga pelos passageiros.

* Interação:
Há também uma interação significativa entre classe e sex. Isso sugere que a diferença nas tarifas entre as classes pode variar dependendo do sexo do passageiro (por exemplo, homens e mulheres podem ter diferenças nas tarifas dependendo da classe em que viajam).

## Post Hoc

Para complementar a análise, podemos realizar um teste post hoc para comparar as diferenças entre as classes e sexos. Neste caso, vamos utilizar o teste de Tukey. Podemos utilizar a função MultiComparison do statsmodels para criar um objeto que representa as comparações múltiplas entre os grupos, e em seguida aplicar o teste de Tukey com a função tukeyhsd.

```python
titanic['classe_sex'] = titanic['classe'].astype(str) + '_' + titanic['sex']
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Criamos uma nova coluna juntando as variaveis independentes, isto é feito porque
# as funções não aceitam entradas bidimensionais só unidimensionais.
titanic['classe_sex'] = titanic['classe'].astype(str) + '_' + titanic['sex']


# Com esta função o endog recebe a variavel dependente, e o groups recebe a variavel independente.
tukey = pairwise_tukeyhsd(endog=titanic[['fare']],
                          groups=titanic[['classe_sex']], # groups = titanic[['classe', 'sex']]
                          alpha=0.05)

# Imprimindo os resultados
print(tukey)
```

Podemos ver que todas as combinações de grupos tiveram diferenças significativas nas médias de fare, exceto para a comparação entre Second_female e Second_male, e também para as comparações envolvendo Third_female e Third_male que não tiveram diferenças significativas.

No geral, o teste post hoc é útil quando a análise ANOVA indica que há diferenças significativas entre as médias, mas não nos diz quais grupos são diferentes uns dos outros. O teste post hoc nos ajuda a identificar exatamente quais grupos são diferentes e em que grau.





**Diferenças Significativas (onde reject = True):**

Há uma diferença significativa nas tarifas entre os seguintes pares de grupos:

* First_female vs. First_male: A diferença média é de -38.90, indicando que as mulheres na primeira classe pagam significativamente menos que os homens.

### Tabela Resumo: Diferenças entre One-Way e Two-Way ANOVA

| Característica           | One-Way ANOVA                     | Two-Way ANOVA                          |
|--------------------------|------------------------------------|----------------------------------------|
| **Número de fatores**    | 1 (ex.: `species`)                | 2 (ex.: `species` e `fertilizer`)      |
| **Número de grupos**     | Duas ou mais categorias           | Duas ou mais categorias para cada fator |
| **Interação analisada**  | Não                               | Sim (interação entre os dois fatores)  |
| **Exemplo de pergunta**  | A média de `sepal_length` varia entre espécies? | A média de `sepal_length` varia entre espécies, fertilizantes ou sua interação? |
